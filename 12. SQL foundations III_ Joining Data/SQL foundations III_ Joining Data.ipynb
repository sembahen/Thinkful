```
#1. How many movies are released for each rating?
```

```
sql= '''
SELECT rating, COUNT(*)
FROM film
GROUP BY rating;
'''
results=engine.execute(sql)
engine.dispose()
rows=results.fetchall()

for row in rows:
  print(row)
```

```
# 2.What is the average rental duration for each rating?
```

```
sql= '''
SELECT rating , AVG(rental_duration)
FROM film
GROUP BY rating;
'''
results=engine.execute(sql)
engine.dispose()
rows=results.fetchall()

for row in rows:
  print(row)
```
