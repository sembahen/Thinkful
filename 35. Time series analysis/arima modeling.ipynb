```
import numpy as np
import pandas as pd
import scipy
from datetime import datetime
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
```

```
x_ar = [15]
for i in range(500):
    y = 5 + .7*x_ar[i-1] 
    x_ar.append(float(y))
```

```
plt.plot(x_ar)
plt.title('Autoregressive Example')
plt.show()
```

```
x_ar = [15]
for i in range(500):
    y = 5 + 1*x_ar[i-1] 
    x_ar.append(float(y))
plt.plot(x_ar)
plt.title('Autoregressive Example')
plt.show()
```

```
#changing the value of   ùõΩ to -1
x_ar = [15]
for i in range(500):
    y = 5 + -1*x_ar[i-1] 
    x_ar.append(float(y))
plt.plot(x_ar)
plt.title('Autoregressive Example')
plt.show()
```
